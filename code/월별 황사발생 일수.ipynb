```python
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
```

```python
mpl.rc('font', family = fm.FontProperties(fname='C:/Windows/Fonts/malgun.ttf').get_name())
plt.style.use('dark_background')
```

```python
monthly_df = pd.read_csv('../data/서울_1993_2022_월별_황사일수.txt', index_col = 0)
monthly_df
```

```python
season_df = pd.read_csv('../data/서울_1993_2022_계절별_황사일수.txt', index_col = 0)
season_df
```

```python
season_mean = season_df.mean()
season_mean
```

```python
fig, sub = plt.subplots(figsize = (6, 3))
bar = sub.bar(season_mean.index, season_mean, color = '#EEEEEE')
sub.bar_label(bar, fmt='%.1f', fontweight = 'bold')
sub.set_xticklabels(season_mean.index, rotation = 0)
sub.set_xlabel('계절')
sub.set_ylabel('평균 발생 일수')
sub.set_ylim(0, 8)
plt.show()
```

```python
fig, sub = plt.subplots(figsize = (6, 3))
for season, c in zip(season_df.columns, ['#D6F1C6', '#C9F1FD', '#F9CC87', '#FEF8FA']):
    sub.plot(season_df[season], c = c, label = season, lw = 2)
sub.legend()
sub.grid(ls = '--', lw = .5)
sub.set_xlabel('연도')
sub.set_ylabel('발생 일수')
plt.show()
```
